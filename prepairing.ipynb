```python
with open('ground_truth.txt', 'r') as file:
    lines = file.readlines()

if len(lines) < 3:
    raise ValueError("Файл содержит менее трех строк.")

rppg = lines[0].strip()
time = lines[2].strip()

rppg = list(map(float, rppg.split()))
time = list(map(float, time.split()))
```

```python
import matplotlib.pyplot as plt

plt.plot(time[:50], rppg[:50], linestyle='-')
plt.show()
```
